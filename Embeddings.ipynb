```python
import os.path
from urllib.request import urlretrieve

base = "https://edwardslab.bmcb.georgetown.edu/~nedwards/dropbox/6ItUS2tEdC/"
csvfile = "pride-embeddings-openai-3-small.csv"
if not os.path.exists(csvfile):
  urlretrieve(base+csvfile,csvfile)
fthfile = "pride-embeddings-openai-3-small.fth"
if not os.path.exists(fthfile):
  urlretrieve(base+fthfile,fthfile)

import pandas as pd

df = pd.read_feather(fthfile)
md = pd.read_csv(csvfile)
print(df)
print(md)
```
