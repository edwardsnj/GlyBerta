```python
couchws.get('PXD005295')
```

```python
# Define a CouchDB query
query = """
{
  "selector": {
    "submissionType": "COMPLETE",
    "organisms": {
      "$elemMatch": {
        "name": "Homo sapiens (human)"
      }
    },
    "identifiedPTMStrings": {
      "$elemMatch": {
        "name" : {
          "$in": ["deamidated residue","Deamidated"]
        }
      }
    },
    "$text": "sampleProcessingProtocol:PNGase"
  }
}
"""
```

```python
pride_ids = [ doc["accession"] for doc in couchws.search(query) ]
len(pride_ids)
```
